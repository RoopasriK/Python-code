1.1 Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce()

```python
def myreduce(func,seq):
    result = seq[0]
    for item in seq[1:]:
        result = func(result,item)  
    return result
#sample test
myreduce(lambda x,y : x if x>y else y,[15,4,35,12,44,10])
```

1.2 Write a Python program to implement your own myfilter() function which works exactly like
Python's built-in function filter()

```python
def myfilter(func,seq):
    result = []
    for item in seq:
        if(func(item)):
            result.append(item)
    return result
#sample to demonstrate
n = list(myfilter(lambda x : True if x%2==0 else False,[10,15,20,25,11,22,33]))
print(n)
```

3 Implement a function longestWord() that takes a list of words and returns the longest one.

```python
entries = input("Enter a list of words : ")
def longestWord(cards):
    list2 = []
    for x in cards:
        list2.append([len(x),x])
    list2.sort(reverse=True)
    if(list2[0][0]!=list2[1][0]):
        return list2[0][1]       
    else:
        return list2[0][1],list2[1][1]
longestWord(list(entries.split(',')))
```

1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
class shape:
    def __init__(self):
        self.side = int(input("Enter the side of triangle "))
        self.len1 = int(input("Enter the length1 of triangle "))
        self.len2 = int(input("Enter the length2 of triangle "))
        self.len3 = int(input("Enter the length3 of triangle "))
class triangle(shape):
    def area(self,*args):
        super(triangle,self).__init__(*args)
    def calc(self):
        return (self.side*(self.side-self.len1)*(self.side-self.len2)*(self.side-self.len3))*0.5
s = triangle()
s.calc()
```

1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n.

```python
entries = input("Enter a list of words : ")
n = int(input("Enter a number : "))
def filter_long_words(words, n):
    return list(filter(lambda x: len(x) > n, words))
filter_long_words(list(entries.split(',')),n)
```
